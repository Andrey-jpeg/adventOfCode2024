```kotlin
import java.io.File

val regex = "mul\\(\\d{1,3},\\d{1,3}\\)".toRegex()

File("day3.txt").readLines().fold(0) { acc, line ->
    val matches = regex.findAll(line)
    
    acc + matches.fold(0) { acc, match ->
        val digitRegex = "\\d+".toRegex()
        val (x, y) = digitRegex.findAll(match.value).toList().map { it.value.toInt() }
        acc + x * y
    }
}
```

```kotlin
val regex = "mul\\(\\d{1,3},\\d{1,3}\\)|don\\'t\\(\\)|do\\(\\)".toRegex()

var enabled = true

File("day3.txt").readLines().fold(0) { acc, line ->
    val matches = regex.findAll(line)
    
    acc + matches.fold(0) { acc, match ->
        if (match.value.equals("don't()")) {
            enabled = false
            return@fold acc + 0
        }

        if (match.value.equals("do()")) {
            enabled = true
            return@fold acc + 0
        }

        if (!enabled) {
            return@fold acc + 0
        }

        val digitRegex = "\\d+".toRegex()
        val (x, y) = digitRegex.findAll(match.value).toList().map { it.value.toInt() }
        acc + x * y
    }
}
```
